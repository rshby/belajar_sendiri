```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('flights')
```

```python
df
```

```python
sns.relplot(x='passengers', y='month', data=df)
```

```python
sns.relplot(x='passengers', y='month', hue='year', data=df)
```

```python
df = sns.load_dataset('tips')
sns.relplot(x='time', y='tip', kind='line', data=df)
```

```python
sns.catplot(x="day", y='total_bill', data=df)
```

```python
sns.catplot(x="day", y='total_bill', kind='boxen', data=df)
```

# Multiple Grid

```python
a = sns.load_dataset('iris')
b = sns.FacetGrid(a, col='species')
b.map(plt.hist, 'sepal_length')
```

```python
a = sns.load_dataset('flights')
b = sns.PairGrid(a)
b.map(plt.scatter)
```

# Plot Aesthetic

```python
sns.set(style='dark')
a = sns.load_dataset('flights')
b = sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style='white', color_codes=True)
a = sns.load_dataset('tips')
sns.boxplot(x='day', y='total_bill', data=a)
```

```python
sns.set(style='white', color_codes=True)
a = sns.load_dataset('tips')
sns.boxplot(x='day', y='total_bill', data=a)
sns.despine(offset=10, trim=True)
```
