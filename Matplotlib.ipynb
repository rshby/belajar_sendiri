- Selain menyertakan duah buah deret bilangan sebagai parameter untuk sebumbu x dan y, terdapat parameter ketiga yang bisa kita sertakan untuk mengatur warna dan jenis plotting.

## Plotting dengan keyword
Matplotlib juga memungkinkan kita untuk melakukan plotting sekumpulan nilai yang tersimpan dalam struktur data yang disertai keywords argument(kwargs) seperti pada `dictionary` dan `Pandas DataFrame`.

```python
data = {
    'a' : np.arange(50),
    'c' : np.random.randint(0, 50, 50),
    'd' : np.random.rand(50)
}

data["b"] = data["a"] + 10*np.random.rand(50)
data["d"] = np.abs(data["d"])*100

data
```

```python
plt.scatter('a', 'b', data=data)
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.show()
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.show()
```

# Histogram

```python
mu, sigma = 100, 15

x = mu + sigma * np.random.randn(10000)
x
```

```python
x.shape
```

```python
plt.hist(x,
        bins=50,
        facecolor='g',
        alpha=0.75)

plt.xlabel("Sumbu x")
plt.ylabel("Sumbu Y")
plt.title("Contoh Histogram")

plt.text(45, 500, r'$\mu=100,\ \sigma=15$')
plt.grid()

plt.show()
```

## Histogram dengan OO Style

```python
fig, ax = plt.subplots()

ax.hist(x,
        bins=50,
        facecolor='r',
        alpha=0.75)

ax.set_xlabel("Sumbu x")
ax.set_ylabel("Sumbu Y")
ax.set_title("Contoh Histogram")

ax.text(45, 500, r'$\mu=100,\ \sigma=15$')
ax.grid()

plt.show()
```

# Bar plot / Bar Chart

```python
data = [25, 45, 55, 125, 225]
kategori = ["A", "B", "C", "D", "E"]
```

```python
plt.figure(figsize=(8,4))
plt.bar(kategori, data)
plt.grid()

plt.xlabel("Kategori")
plt.ylabel("Jumlah")
plt.title("Contoh Bar Plot")

plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.bar(
    kategori,
    data,
    color='g',
    alpha=0.35
)

plt.grid(linestyle='--',
        linewidth=2,
        axis='y', # pilihan: x, y, both
        alpha=0.80)

plt.xlabel("Kategori")
plt.ylabel("Jumlah")
plt.title("Contoh Bar Plot")

plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.barh(kategori[::-1], data[::-1], color='b', alpha=0.20)
plt.grid(linestyle='--',
        linewidth=2,
        axis='x', # pilihan: x, y, both
        alpha=0.90)

plt.xlabel("Jumlah")
plt.ylabel("Kategori")
plt.title("Contoh Bar Plot")

plt.show()
```

## Stack Bar Plot

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ["A", "B", "C", "D", "E"]
```

```python
plt.figure(figsize=(8,4))
plt.bar(kategori, data1, label="Data 1")
#plt.bar(kategori, data2, label="Data 2")
plt.bar(kategori, data2, label="Data 2", bottom=data1)

plt.legend()

plt.grid(linestyle='--',
        linewidth=1,
        axis='y', # pilihan: x, y, both
        alpha=0.90)

plt.xlabel("Kategori")
plt.ylabel("Jumlah")
plt.title("Contoh Bar Plot")

plt.show()
```

## Group Bar Plot

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ["A", "B", "C", "D", "E"]

x = np.arange(len(kategori))
x
```

```python
width = 0.38

plt.bar(x-width/2, data1, width, label="Data 1")
plt.bar(x+width/2, data2, width, label="Data 2")

plt.xticks(x, kategori)

plt.grid(linestyle='--',
        linewidth=1,
        axis='y', # pilihan: x, y, both
        alpha=0.90)

plt.xlabel("Kategori")
plt.ylabel("Jumlah")
plt.title("Contoh Bar Plot")

plt.legend()
plt.show()
```

# Pie Plot / Pie Chart

```python
data = [215, 130, 245, 210, 225]
kategori = ["A", "B", "C", "D", "E"]
```

```python
plt.figure(figsize=(6, 6))
plt.pie(data,
       labels=kategori,
       autopct='%1.2f%%',
       startangle=90)

plt.title("Contoh Pie Plot")

plt.show()
```

## Pengaturan Legend dan Color

```python
data = [215, 130, 245, 210, 225]
kategori = ["A", "B", "C", "D", "E"]
warna = ["pink", "cyan", "green", "yellowgreen", "skyblue"]
```

```python
plt.figure(figsize=(6, 6))
plt.pie(data,
       labels=kategori,
       colors=warna,
       autopct="%1.2f%%",
       startangle=0)

plt.legend()
plt.title("Contoh Pie Plot")

plt.show()
```

## Exploded Pie Plot

```python
data = [215, 130, 245, 210, 225]
kategori = ["A", "B", "C", "D", "E"]
warna = ["pink", "cyan", "green", "yellowgreen", "skyblue"]
explode_var = [0., 0., 0.12, 0., 0.08]
```

```python
plt.figure(figsize=(6, 6))
plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct="%1.2f%%",
        startangle=90)

plt.legend()
plt.title("Contoh Pie Plot")

plt.show()
```

# Table Plot / Table Chart

```python
data_uas = [
    ["Bejo", 70],
    ["Tejo", 83],
    ["Cecep", 90],
    ["Wati" , 74],
    ["Reo", 100]
]
```

```python
data_uas
```

### Table plot dengan OO Style

```python
fig, ax = plt.subplots()

table = plt.table(cellText=data_uas,
                 loc='center')

table.set_fontsize(14)
table.scale(0.7, 3) # ukuran kolom, baris

ax.axis(False)

plt.show()
```

### Table Plot dengan pyplot style

```python
table = plt.table(cellText=data_uas,
                 loc='center')

table.set_fontsize(14)
table.scale(1, 4)

ax = plt.gca()
ax.axis(False)

plt.show()
```
