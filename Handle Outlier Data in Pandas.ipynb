```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Datasets

```python
df = pd.read_csv("datasets/weight-height.csv")
df.head(10)
```

```python
df.shape
```

### Buat Histogram Sebelum Data Outlier dihandle

```python
fig, axes = plt.subplots(2, 1, figsize=(15, 10))

fig.suptitle('Histogram Sebelum di Handle')

sns.distplot(df["Height"], ax=axes[0])
sns.distplot(df["Weight"], ax=axes[1], color='r')

plt.show()
```

### Menangani data Height

```python
Q1 = df["Height"].quantile(0.25)
Q3 = df["Height"].quantile(0.75)
print(f"Batas Q1: {Q1} || Batas Q3: {Q3}")
```

```python
# Menghitung IQR
IQR = Q3-Q1
IQR
```

```python
# set lower dan upper
low_height = Q1 - (1.5*IQR)
up_height = Q3 + (1.5*IQR)
print(f"Lower: {low_height} || Upper: {up_height}")
```

```python
filt = (df["Height"]>low_height) & (df["Height"]<up_height)
df = df.loc[filt]
```

```python
df.shape
```

### Menangani Kolom Weight

```python
Q1_Weight = df["Weight"].quantile(0.25)
Q3_Weight = df["Weight"].quantile(0.75)
print(f"Batas Q1 Weight: {Q1_Weight}\nBatas Q3 Weight: {Q3_Weight}")
```

```python
# Menghitung IQR
IQR_Weight = Q3_Weight-Q1_Weight
print(f"IQR Weight: {IQR_Weight}")
```

```python
# set lower dan upper
low_weight = Q1_Weight - (1.5*IQR_Weight)
up_weight = Q3_Weight + (1.5*IQR_Weight)
print(f"Lower Weight: {low_weight}\nUpper Weight: {up_weight}")
```

```python
filt_weight = (df["Weight"]>low_weight) & (df["Weight"]<up_weight)
df = df.loc[filt_weight]
df.shape
```

```python
df.describe()
```

### Histogram Setelah data di handle

```python
fig, axes = plt.subplots(2, 1, figsize=(15, 10))

fig.suptitle('Histogram Sebelum di Handle')

sns.distplot(df["Height"], ax=axes[0])
sns.distplot(df["Weight"], ax=axes[1], color='r')

plt.show()
```
